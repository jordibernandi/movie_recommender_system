```python
import numpy as np 
import pandas as pd
```

```python
movies = pd.read_csv('data/netflix_titles.csv')
```

```python
movies.head(2)
```

```python
movies = movies[['title','director','cast','country','date_added','release_year', 'listed_in', 'description']]
```

```python
movies.head(2)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies = movies.apply(lambda x: x.map(lambda v: v.lower() if isinstance(v, str) else v))
```

```python
movies.head(2)
```

```python
def split_str(x):
    if isinstance(x, str):
        return x.split(",")
    else:
        return []
```

```python
movies['director'] = movies['director'].apply(split_str)
movies['cast'] = movies['cast'].apply(split_str)
movies['country'] = movies['country'].apply(split_str)
movies['listed_in'] = movies['listed_in'].apply(split_str)
```

```python
movies.head(2)
```

```python
def remove_space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
def process_title(s):
    result = s.split(":")[0].strip()
    result = result.split("-")[0].strip()
    return result
```

```python
print(process_title("Go! Go! Cory Carson: Chrissy Takes the Wheel"))
```

```python
movies['title'] = movies['title'].apply(process_title)
```
