```python
import pandas as pd
import os
```

```python
# Read the CSV files
netflix_history_df = pd.read_csv('../data/netflix_history.csv')
```

```python
netflix_history_df
```

```python
netflix_history_df['Title'] = netflix_history_df['Title'].str.lower()
```

```python
len(netflix_history_df['Title'].unique())
```

```python
# Split the text in the 'Title' column by ':'
title_parts = netflix_history_df['Title'].str.split(':')
netflix_history_df['Title'] = title_parts.str[0]

# Split the text in the 'Title' column by '-'
title_parts = netflix_history_df['Title'].str.split('–')
netflix_history_df['Title'] = title_parts.str[0]
```

```python
len(netflix_history_df['Title'].unique())
```

```python
netflix_history_df['Start Time'] = pd.to_datetime(netflix_history_df['Start Time'], dayfirst=True)
netflix_history_df['Duration'] = pd.to_timedelta(netflix_history_df['Duration'])
```

```python
netflix_history_df['Duration_seconds'] = netflix_history_df['Duration'].dt.total_seconds()
```

```python
filtered_df = netflix_history_df[netflix_history_df['Attributes'].isna()]
```

```python
filtered_df['Duration'].isna().sum()
```

```python
filtered_df
```

```python
# Sum the 'Duration_seconds' based on the same 'Title'
sum_duration_by_title = filtered_df.groupby('Title')['Duration_seconds'].sum()

# Filter out titles with a total duration shorter than 30 minutes (1800 seconds)
filtered_titles = sum_duration_by_title[sum_duration_by_title >= 1800].index
```

```python
# Keep only one row of data for each 'Title' based on the earliest 'Start Time'
filtered_df = filtered_df[filtered_df['Title'].isin(filtered_titles)]
filtered_df = filtered_df.sort_values(by=['Start Time']).groupby('Title').first().reset_index()
```

```python
# Add a column noting the total duration for each title
filtered_df = pd.merge(filtered_df, sum_duration_by_title.reset_index(name='Total Duration'), on='Title', how='left')
filtered_df = filtered_df.sort_values(by=['Start Time']).reset_index()
```

```python
filtered_df = filtered_df[['Title', 'Start Time', 'Total Duration']]
```

```python
filtered_df
```

```python
filtered_df.to_csv('../data/netflix_history_preprocessed.csv', index=False)
```
